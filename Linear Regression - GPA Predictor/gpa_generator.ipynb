GPA Generator - Generates dataset of a uniform distribution of scores for courses specified alongside the GPA corresponding to the scores.

```python
import pandas as pd
import random
```

Setting of subjects and their corresponding units

```python
subjects = ['CHE 201', 'MTH 201', 'MEE 203', 'MEE 205', 'CSC 201', 'EEE 203', 'EEE 291', 'MSE 201']
course_units = [3, 4, 2, 3, 3, 3, 1, 3]
courses_df = pd.DataFrame(columns=subjects)
```

Setting of dataset length and maximum and minimum scores

```python
dataset_number = int(input("Enter the dataset length: "))
mn, mx = int(input("Enter the minimum score: ")), int(input("Enter the maximum score: "))
for i in range(dataset_number):
    scores_list = []
    for j in range(len(subjects)):
        scores_list.append(random.randint(mn, mx))
    courses_df.loc[i] = scores_list
```

```python
def gpa_calc(scores_list, course_units):
    summation = 0
    for i in range(len(scores_list)):
        if scores_list[i] >= 70:
            summation += 5 * course_units[i]
        elif scores_list[i] >= 60:
            summation += 4 * course_units[i]
        elif scores_list[i] >= 50:
            summation += 3 * course_units[i]
        elif scores_list[i] >= 45:
            summation += 2 * course_units[i]
        elif scores_list[i] >= 40:
            summation += 1 * course_units[i]
        else:
            summation += 0

    return summation / sum(course_units)

gpa = []
for i in range(dataset_number):
    scores_list = list(courses_df.loc[i]) 
    gpa.append(gpa_calc(scores_list, course_units))

gpa
```

```python
courses_df['gpa'] = gpa
courses_df
```

```python

courses_df.to_csv(f'gpa_{input("Enter dataset identifier: ")}_scores.csv')
```
